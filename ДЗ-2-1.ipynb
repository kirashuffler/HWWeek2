# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql 
select tracks.Name  
from tracks join invoice_items 
on tracks.TrackId = invoice_items.TrackId 
join invoices on invoice_items.InvoiceId = invoices.InvoiceId 
join customers on customers.CustomerId = invoices.InvoiceId 
where customers.CustomerId = 
(select customers.CustomerId 
from customers join employees 
on customers.SupportRepId = employees.EmployeeId 
where employees.HireDate > '2002-09-01')
```

## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
select playlists.Name, round(sum(tracks.UnitPrice),2), sum(tracks.Milliseconds) from 
playlists left join playlist_track on playlists.PlaylistId = playlist_track.PlaylistId
left join tracks on tracks.TrackId = playlist_track.TrackId
group by playlists.Name
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql
select artists.Name, genres.Name, count(tracks.GenreId) from 
artists join albums on artists.ArtistId = albums.ArtistId
join tracks on albums.AlbumId = tracks.AlbumId
join genres on genres.GenreId = tracks.GenreId
group by artists.Name
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql
select customers.Company, count(tracks.Name)
from tracks join genres on tracks.GenreId = genres.GenreId 
join invoice_items on tracks.TrackId = invoice_items.TrackId 
join invoices on invoice_items.InvoiceId = invoices.InvoiceId 
join customers on customers.CustomerId = invoices.InvoiceId
where genres.Name = 'Rock'
group by customers.Company
```
